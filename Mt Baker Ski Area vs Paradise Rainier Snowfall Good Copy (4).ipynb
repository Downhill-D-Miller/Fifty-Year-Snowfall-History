Now that two lists consisting of 50 elements, with each element as the annual recorded snowfall for a given season, a frequency table along with tables displaying the proportions and percentages of years having snowfall totals within a given range can be calculated. The snowfall ranges were abritarily selected since what one considers a high annual total snowpack year can be based on subjective opinion. Another method to acquire ranges of snowfall depth and plot them into a pie chart is to calculate the standard deviation and then group the seasons accordingly. For example: Seasons one standard deviation from the mean could be classified as a 'low' snow year, two standard deviations from the mean a 'very low' snow year. Likewise, seasons with one standard deviation above the mean can be considered as 'high' snow years and two standard deviations as 'very high' and so forth. One could calculate what percentage of seasons were one, two standard deviations from the mean in either direction.

In creating the function for these frequency tables, I begin by assigning an input variable that will enable one to use the function with any related dataset. For example if one wanted to do the past 50 years of annual snowfall for another location. Next, one will name their keys as they choose and then having the value set to zero. The frequency of proportions along with the frequency of percentages will start off being assigned an empty dictionary. The next step is to run a for loop using the input value in your function, in this case, 'dataset'. You will then create control flow using a series of comparison operators to determine how many seasons fit in a given range of snowfall totals. These values are abritrary. 

In order to figure out proportion and percentage frequencies, you will run a for loop using your frequency table to reasign the 'prop_table' and 'percent_table' dictionaries their new values. The function will return three frequency tables in list format(frequency of annual snowfall totals falling with in a certain range, the proportion of years over the past 50 seasons having snowfall totals within a given range, and then multiplying the proportions by 100 to determine the percentages. 

```python
def freq_tables(dataset):
    frequency_table = {'Low': 0, 'Medium': 0, 'High': 0, 'Ultimate': 0}
    prop_table = {}
    percent_table = {}
    for data in dataset:
        if data < 400:
            frequency_table['Low'] += 1
        elif 400 <= data < 600:
            frequency_table['Medium'] += 1
        elif 600 <= data < 800:
            frequency_table['High'] += 1
        elif data >= 800:
            frequency_table['Ultimate'] +=1
    for values in frequency_table:
        proportion = frequency_table[values] / len(dataset)
        percent = proportion * 100
        prop_table[values] = proportion
        percent_table[values] = percent
    return[frequency_table, prop_table, percent_table]
        
    
```

* Step 1: Pass in the selected list into the function as an argument and assign the results to a variable

* Step 3: Extract the values from the first frequncy table(index positon is 0), and assign the values to a variable
* Step 4: Extract the keys from this first frequency table(index position is 0), and assign the keys to a variable
* Step 5: Create a pie chart using the the two variables you created in step 3 and in step 4. The first variable is a of the 'dict_values' class and will be displayed as the percentages. The second variable which is of the 'dict_keys' class will be used as labels when creating the pie chart.

```python
baker_tables = freq_tables(baker_50_year_snowfall)
print(baker_tables)
baker_table_values = baker_tables[0].values()
baker_table_keys = baker_tables[0].keys()
print(baker_table_values)
print(baker_table_keys)
print(type(baker_table_values))
print(type(baker_table_keys))
```

To get a 50-year year-by-year comparison of snowfall totals for the two different locations, you can use a line chart. Given the two list generated earlier depicting the annual snowfall for each season, those lists can used as values on the y-axis of the line graph. We will have the the 50 seasons on the x-axis. To save from the hassle of having to type all 50 years in creating a list, defining a function using the range function within the function can be used. 

```python
# Generating the list of each of the seasons starting with 1970 and finishing with 2019. 

def fifty_years(r1, r2):
    return list(range(r1, r2 + 1))
```

```python
fifty_year_list = fifty_years(1970, 2019) # Call our function and assign the starting and ending value and assign to a variable
print(fifty_year_list)
```
